```python
    
def plot_vality_params(index, data, name, x_label, y_label, title=None):
    plt.figure(figsize=(7,5))
    for i, (index, value) in enumerate(zip(index, data)):
        value = value * 100
        print(index, value)
        Index = ["One_hot + Position", "1.9M(MolGPT)", "6M(MolGPT)"]
        if index in Index:
            
            plt.bar(index, value, width=0.7, color='orange')
            

        else:
            
            plt.bar(index, value, width=0.7, color = "navy")
    plt.ylim(70, 100)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(name)
    plt.show()
```

```python
plot_vality_params(datasize_index, vality, "Vality Rate", "Encoding Method", "Validity")
```
